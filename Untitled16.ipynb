# Model Evaluation and Visualization (Machine Learning):

This section evaluates different machine learning models using cross-validation, metrics calculation, and visualization.
It includes logistic regression, random forest, decision tree, XGBoost, SVM, and various metrics and visualizations.

```python
from sklearn.svm import SVC

# Models to evaluate
models = [
    LogisticRegression(),
    RandomForestClassifier(),
    DecisionTreeClassifier(),
    XGBClassifier(),
]

# Initialize dictionary to store metrics for each model
metrics_data = {'Model': [], 'Accuracy': [], 'Precision': [], 'Recall': [], 'F1-Score': []}

for model in models:
    # Cross-Validation
    cv_scores = cross_val_score(model, X_train_scaled, y_train, cv=5, scoring='accuracy')
    print(f"\n{type(model).__name__} Cross-Validation Scores: {cv_scores}")
    print(f"Mean CV Accuracy: {np.mean(cv_scores):.4f}")

    # Model Training
    model.fit(X_train_scaled, y_train)

    # Model Prediction
    predictions = model.predict(X_test_scaled)
    probabilities = model.predict_proba(X_test_scaled)[:, 1]
    model_name = type(model).__name__

    # Calculate metrics
    accuracy = accuracy_score(y_test, predictions)
    precision = precision_score(y_test, predictions, average=None)
    recall = recall_score(y_test, predictions, average=None)
    f1 = f1_score(y_test, predictions, average=None)

    # Display and store metrics
    print(f"\n{model_name} Test Metrics:")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision}")
    print(f"Recall: {recall}")
    print(f"F1-Score: {f1}")

    # Classification Report
    print(f"\nClassification Report for {model_name}:")
    print(classification_report(y_test, predictions))

    # Store metrics in the dictionary
    metrics_data['Model'].append(model_name)
    metrics_data['Accuracy'].append(accuracy)
    metrics_data['Precision'].append(precision)
    metrics_data['Recall'].append(recall)
    metrics_data['F1-Score'].append(f1)

    # Confusion Matrix
    cm = confusion_matrix(y_test, predictions)
    plt.figure(figsize=(4, 3))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(f'{model_name} Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

    # Precision-Recall curve
    precision, recall, _ = precision_recall_curve(y_test, probabilities)
    pr_auc = auc(recall, precision)
    plt.figure(figsize=(4, 3))
    plt.plot(recall, precision, color='darkorange', lw=2, label=f'{model_name} (AUC = {pr_auc:.2f})')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title(f'{model_name} Precision-Recall Curve')
    plt.legend(loc='upper right')
    plt.show()

# Convert metrics_data to a DataFrame for easy plotting
metrics_df = pd.DataFrame(metrics_data)

# Calculate and display average metrics for each model
for model_name in metrics_data['Model']:
    model_metrics = metrics_df[metrics_df['Model'] == model_name].iloc[:, 1:]
    avg_precision = model_metrics['Precision'].apply(np.mean)
    avg_recall = model_metrics['Recall'].apply(np.mean)
    avg_f1_score = model_metrics['F1-Score'].apply(np.mean)

    print(f"\nAverage metrics for {model_name}:")
    print(f"Average Precision: {avg_precision}")
    print(f"Average Recall: {avg_recall}")
    print(f"Average F1-Score: {avg_f1_score}")
```

# Model Evaluation and Visualization (Deep Learning):

Similar to the previous block, but for deep learning models.
It includes evaluation of artificial neural networks (ANNs), multi-layer perceptrons (MLPs), and convolutional neural networks (CNNs).

```python

# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Create a DataFrame to store the results
results_df = pd.DataFrame(columns=['Model', 'Accuracy', 'Precision', 'Recall', 'F1 Score'])

# Function to evaluate and append results to the DataFrame
def evaluate_and_append_results(model, X_train, y_train, X_test, y_test, model_name):
    # Train the model
    model.fit(X_train, y_train, epochs=25, batch_size=64, validation_split=0.2, verbose=0)

    # Evaluate the model
    predictions = (model.predict(X_test) > 0.5).astype(int).flatten()

    # Calculate metrics
    accuracy = accuracy_score(y_test, predictions)
    precision = precision_score(y_test, predictions)
    recall = recall_score(y_test, predictions)
    f1 = f1_score(y_test, predictions)

    # Append results to the DataFrame
    results_df.loc[len(results_df)] = [model_name, accuracy, precision, recall, f1]

    # Display Test Accuracy and Classification Report
    print(f"\n{model_name} Test Accuracy: {accuracy:.4f}")
    print(f"{model_name} Classification Report:\n", classification_report(y_test, predictions))

    # Plot confusion matrix and metrics
    plt.figure(figsize=(12, 4))

    metrics_names = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
    metrics_values = [accuracy, precision, recall, f1]

    plt.subplot(1, 2, 2)
    plt.bar(metrics_names, metrics_values, color=['skyblue', 'orange', 'green', 'red'])
    plt.title(f'{model_name} Metrics')
    plt.ylim(0, 1)

    plt.tight_layout()
    plt.show()

# Build and evaluate the ANN model
ann_model = Sequential([
    Dense(64, input_dim=X_train_scaled.shape[1], activation='relu'),
    Dense(1, activation='sigmoid')
])
ann_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
evaluate_and_append_results(ann_model, X_train_scaled, y_train, X_test_scaled, y_test, "ANN")

# Build and evaluate the MLP model
mlp_model = Sequential([
    Dense(64, input_dim=X_train_scaled.shape[1], activation='relu'),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])
mlp_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
evaluate_and_append_results(mlp_model, X_train_scaled, y_train, X_test_scaled, y_test, "MLP")

# Build and evaluate the CNN model
X_train_cnn = X_train_scaled.reshape(X_train_scaled.shape[0], X_train_scaled.shape[1], 1)
X_test_cnn = X_test_scaled.reshape(X_test_scaled.shape[0], X_test_scaled.shape[1], 1)

cnn_model = Sequential([
    Conv1D(32, 3, activation='relu', input_shape=(X_train_cnn.shape[1], 1)),
    MaxPooling1D(2),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')
])
cnn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
evaluate_and_append_results(cnn_model, X_train_cnn, y_train, X_test_cnn, y_test, "CNN")

# Print the DataFrame
print(results_df)

# Calculate and display average metrics for each model
avg_metrics = results_df.groupby('Model').mean()
print("\nAverage Metrics:")
print(avg_metrics)

# Plot results
plt.figure(figsize=(8, 4))
metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
for metric in metrics:
    sns.barplot(x='Model', y=metric, data=results_df)
    plt.title(f'Comparison of {metric} for Different Models')
    plt.ylabel(metric)
    plt.show()
```

# Comparison of Models:

This block compares the performance of different models (both ML and DL) based on various evaluation metrics.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample ML and DL results
ml_results = {
    'Model': ['Logistic Regression', 'Random Forest Classifier', 'Decision Tree Classifier', 'XGB Classifier', 'SVM'],
    'Accuracy': [0.72, 0.81, 0.73, 0.76, 0.72],
    'Precision': [0.69, 0.79, 0.70, 0.74, 0.70],
    'Recall': [0.68, 0.79, 0.70, 0.74, 0.69],
    'F1-Score': [0.68, 0.79, 0.70, 0.74, 0.69]
}

dl_results = {
    'Model': ['ANN', 'MLP', 'CNN'],
    'Accuracy': [0.75, 0.74, 0.74],
    'Precision': [0.66, 0.65, 0.69],
    'Recall': [0.64, 0.62, 0.53],
    'F1-Score': [0.65, 0.63, 0.60]
}

# Convert dictionaries to DataFrames
ml_df = pd.DataFrame(ml_results)
dl_df = pd.DataFrame(dl_results)

# Melt the DataFrame for easier plotting
ml_melted_df = pd.melt(ml_df, id_vars=['Model'], var_name='Metric', value_name='Score')
dl_melted_df = pd.melt(dl_df, id_vars=['Model'], var_name='Metric', value_name='Score')

# Concatenate ML and DL DataFrames
combined_df = pd.concat([ml_melted_df, dl_melted_df], ignore_index=True)

# Plotting with enhancements
plt.figure(figsize=(12, 6))

# Using 'viridis' palette with solid colors
colors = sns.color_palette("viridis", 4)

# Using solid colors from the 'viridis' palette
sns.barplot(x='Model', y='Score', hue='Metric', data=combined_df, palette=colors, edgecolor='grey')

# Adding grid lines
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adding annotations
for p in plt.gca().patches:
    plt.gca().annotate(f"{p.get_height():.2f}", (p.get_x() + p.get_width() / 2., p.get_height()),
                       ha='center', va='center', fontsize=7, color='black', xytext=(0, 5),
                       textcoords='offset points')

plt.title('Comparison of Metrics for ML and DL Models', fontsize=16)
plt.xlabel('Model', fontsize=12)
plt.ylabel('Score', fontsize=7)
plt.legend(title='Metric', fontsize=10, title_fontsize=12, loc='lower left')
plt.xticks(rotation=45, ha='right', fontsize=7)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```
